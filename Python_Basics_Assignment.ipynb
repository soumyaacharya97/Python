##**Q1.  What is Python and why is it widely used in Data Analytics? Mention any three reasons.**

##**A-**

Python is a popular, high-level, interpreted programming language known for its simple, readable syntax, making it accessible for beginners. It is widely used in data analytics because of its rich ecosystem of libraries, ease of learning, and versatility in managing complex data manipulation.

**Three Reasons Python is Widely Used in Data Analytics:**

**1.Extensive Data Science Libraries:**
Python offers powerful, specialized libraries—such as Pandas for data manipulation, NumPy for numerical computing, and Matplotlib/Seaborn for visualization—that simplify complex analysis tasks.

**2.Simple and Easy-to-Learn Syntax:**
 Its readable, close-to-English syntax reduces the learning curve for beginners and increases productivity, allowing analysts to focus on data insights rather than complex coding.

**3.Versatility and Active Community:**
Python is flexible enough to handle the entire data analysis pipeline—from cleaning and exploration to machine learning and reporting—supported by a large, active, open-source community that constantly improves its tools.

##**Q3. What is a function in Python? Why are functions useful?**

##**A-**

A function in Python is a self-contained, named block of code designed to perform a specific task. It only runs when it is explicitly "called" or invoked, allowing you to execute the same logic multiple times throughout a program without rewriting it.

**Key Benefits of Functions**

**Code Reusability**:
Instead of rewriting the same logic, you can call a function whenever needed, saving time and reducing the total amount of code [1, 25, 29].

**Improved Readability**: Functions allow you to "name" a block of code (e.g., calculate_tax()), which makes the program easier for humans to read and understand [1, 17, 33].

**Easier Maintenance & Debugging**: If you need to fix a bug or change logic, you only have to update it in one place (the function definition) rather than every spot where that code is used [11, 21, 26].

**Abstraction**: You can use a function by knowing its name and what it does without needing to understand its internal complexity [19, 33, 35].

**Modularity**: Functions let you break down a large, complex problem into smaller, independent tasks that can be tested and developed separately [29, 31, 33].

**Enhanced Collaboration**: Different team members can work on separate functions simultaneously without interfering with each other's code [1, 21, 25].

```python
## Q8. Write a Python program to remove duplicate values from a list using a set.
## A-


# Create a list with duplicate values
numbers = [1, 2, 2, 3, 4, 4, 4, 5]

# Convert the list to a set to remove duplicates, then back to a list
unique_numbers = list(set(numbers))

print(f"Original list: {numbers}")
print(f"List after removing duplicates: {unique_numbers}")



```
