Q3. Replace missing values in the Quantity and Price columns with their respective mean values using Pandas.

```python
# Calculate mean values
quantity_mean = df["Quantity"].mean()
price_mean = df["Price"].mean()

# Replace missing values in place
df["Quantity"] = df["Quantity"].fillna(quantity_mean)
df["Price"] = df["Price"].fillna(price_mean)
```

Q4. Remove all rows where the Product Category or Region column contains missing values.

```python
# Remove rows with missing values in specific columns
df = df.dropna(subset=["Product Category", "Region"])

df.isnull().sum()
```

Q5.  Create a new column called Revenue using NumPy, where:

Revenue = Quantity × Price

Display the updated dataset with the new column.

```python

# Create the Revenue column using NumPy multiplication
df["Revenue"] = np.multiply(df["Quantity"], df["Price"])

# Display the updated dataset
print(df)
```

Q7.  Group the dataset by Product Category and calculate the total revenue generated by each category.

```python
# Group by category and sum the revenue
category_revenue = df.groupby("Product Category")["Revenue"].sum().reset_index()

# Display the results
print(category_revenue)
```

Q8. Based on the revenue calculated for each product category, identify the top 3 categories with the
highest revenue and the bottom 3 categories with the lowest revenue.

```python

category_revenue = df.groupby("Product Category")["Revenue"].sum().reset_index()

top_3_categories = category_revenue.sort_values(by="Revenue", ascending=False).head(3)

bottom_3_categories = category_revenue.sort_values(by="Revenue", ascending=True).head(3)

print("--- Top 3 Highest Revenue Categories ---")
print(top_3_categories)

print("\n--- Bottom 3 Lowest Revenue Categories ---")
print(bottom_3_categories)
```

Q9. Group the dataset by Region and calculate the total revenue generated in each region. Identify the
region with the highest revenue and the region with the lowest revenue.

```python
# 1. Group by Region and sum the Revenue
region_revenue = df.groupby("Region")["Revenue"].sum()

# 2. Identify the highest and lowest revenue regions
highest_region = region_revenue.idxmax()
highest_value = region_revenue.max()

lowest_region = region_revenue.idxmin()
lowest_value = region_revenue.min()

# Display results
print("--- Total Revenue by Region ---")
print(region_revenue.reset_index())

print(f"\nHighest Revenue Region: {highest_region} ({highest_value:,.2f})")
print(f"Lowest Revenue Region: {lowest_region} ({lowest_value:,.2f})")
```

Q10.  Extract the month from the Date column using Pandas and group the dataset by month to calculate the
total monthly revenue. Also, calculate the mean, median, and standard deviation of the Revenue column
using NumPy

```python
# 1. Convert Date column to datetime format and extract the month name
df["Date"] = pd.to_datetime(df["Date"])
df["Month"] = df["Date"].dt.strftime("%B")  # Extracts full month name

# 2. Group by Month and calculate total revenue
# Note: You can add sort=False to keep chronological order if your data is already ordered
monthly_revenue = df.groupby("Month")["Revenue"].sum().reset_index()

# 3. Calculate overall summary statistics using NumPy
rev_mean = np.mean(df["Revenue"])
rev_median = np.median(df["Revenue"])
rev_std = np.std(df["Revenue"])

# --- Display Results ---
print("--- Total Monthly Revenue ---")
print(monthly_revenue)

print("\n--- Revenue Overall Summary Statistics ---")
print(f"Mean:               ${rev_mean:,.2f}")
print(f"Median:             ${rev_median:,.2f}")
print(f"Standard Deviation: ${rev_std:,.2f}")
```
