**Q6. Identify any anomalies where houses have high prices but low areas.**

```python
# Define thresholds using the 75th percentile for price and 25th percentile for area
high_price_threshold = df['Price'].quantile(0.75)
low_area_threshold = df['living area'].quantile(0.25)

# Filter rows meeting BOTH anomaly conditions
anomalies = df[(df['Price'] > high_price_threshold) & (df['living area'] < low_area_threshold)]

# Sort them by price (ascending or descending) to see the most extreme cases
anomalies_sorted = anomalies.sort_values(by='Price', ascending=False)

# View the clean anomalies (using .head() or printing the full dataset as preferred)
print(f"Found {len(anomalies_sorted)} structural anomalies.")
print(anomalies_sorted[['Price', 'living area']].head())
```

**Q7. Compare average price based on number of floors and houses with or without waterfront.**

```python
pivot_comparison = df.pivot_table(
    values='Price',
    index='number of floors',
    columns='waterfront present',
    aggfunc='mean')

print(pivot_comparison)
```
